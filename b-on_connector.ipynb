# B-ON connector Notebook

## Prerequisites

1. Add `certificates` folder

2. Configure the following values

```python
B_ON_MQTT_ENDPOINT = ""
TENANT = ""
THING_IMEI = ""
```

## Client initialisation

```python
import paho.mqtt.client as mqtt
import ssl

PATH_TO_CERTIFICATE = "certificates/thing-certificate.pem.crt"
PATH_TO_PRIVATE_KEY = "certificates/thing-private.pem.key"
PATH_TO_ROOT_CA_CERT = "certificates/root-ca.pem"

client = mqtt.Client(
    callback_api_version = mqtt.CallbackAPIVersion.VERSION2,    
    protocol = mqtt.MQTTv5,
    client_id = THING_IMEI,
)
client.tls_set(
    ca_certs = PATH_TO_ROOT_CA_CERT,
    certfile = PATH_TO_CERTIFICATE,
    keyfile = PATH_TO_PRIVATE_KEY, 
    cert_reqs = ssl.CERT_REQUIRED,
    tls_version = ssl.PROTOCOL_TLSv1_2,
    ciphers = None,
)
```

## Sending vehicle data

### Vehicle location

```python
POSITION_LATITUDE = 47.2902794
POSITION_LONGITUDE = 7.7158736
```

```python
from sts.v2.vehicle.signal_pb2 import Signal, Position
from google.protobuf.timestamp_pb2 import Timestamp

mqtt_topic = f"{TENANT}/vehicleStatus/{THING_IMEI}/vehicle/positionStatus"
timestamp = Timestamp()
timestamp.GetCurrentTime()
vehicle_signal = Signal(
    timestamp = timestamp,
    position = Position(lat=POSITION_LATITUDE, lon=POSITION_LONGITUDE),
)

print("Connecting to client")
result = client.connect(host=B_ON_MQTT_ENDPOINT, port=8883)

print("Publishing vehicle location")
result = client.publish(
    topic = mqtt_topic,
    payload = vehicle_signal.SerializeToString(),
    qos = 1,
)

import time
time.sleep(5)

print("Disconnecting from client")
result = client.disconnect()
```
